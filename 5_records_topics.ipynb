```python
squares_ids = list(df_squares.square_contain_id.value_counts().index)
```

```python
pd.read_csv(f'total_squares_without_ad_with_artm_topics/111771.0.csv', sep=';', dtype='str')
```

```python
ad_records_df = pd.DataFrame()

for square_id in squares_ids:
    try:
        df = pd.read_csv(f'total_squares_without_ad_with_artm_topics/{square_id}.0.csv', sep=';', usecols=['id', 'label'], dtype='str')
        lat_lon_df = df_squares[['latitude', 'longitude']].loc[square_id].head(1)

        df['latitude'] = lat_lon_df['latitude'].tolist() * df.shape[0]
        df['longitude'] = lat_lon_df['longitude'].tolist() * df.shape[0]

        ad_records_df = ad_records_df.append(df, ignore_index=True)
    except:
        continue
```
